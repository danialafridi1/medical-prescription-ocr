```python
import os
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
import pytesseract
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR/tesseract.exe'  # Adjust if different
```

```python
# Define dataset paths
data_dir = "../dataset/"  # Update as needed
folders = ["Training", "Testing", "Validation"]
image_size = (250, 250)
```

```python
def load_data(folder):
    
    csv_path = os.path.join(data_dir, folder, "labels.csv")
    image_folder = os.path.join(data_dir, folder, "images")  # Ensure the correct folder

    # Load CSV file
    df = pd.read_csv(csv_path)

    images = []
    labels = []

    for index, row in df.iterrows():
        image_filename = row["IMAGE"]  # Image name from CSV
        image_path = os.path.join(image_folder, image_filename)  # Full image path
        
        if os.path.exists(image_path):  # Ensure image exists
            image = cv2.imread(image_path)
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for matplotlib
            image = cv2.resize(image, (224, 224))  # Resize images if needed
            
            images.append(image)
            labels.append(row["MEDICINE_NAME"])  # Assign label correctly
        else:
            print(f"⚠️ Warning: Image {image_filename} not found in {image_folder}!")

    return np.array(images), np.array(labels)
```

```python
train_images, train_labels = load_data("Training")
test_images, test_labels = load_data("Testing")
val_images, val_labels = load_data("Validation")
```

```python

import cv2
import matplotlib.pyplot as plt
import pytesseract

def segment_characters(image):
    """
    Segments characters from the given image using contour detection.

    Args:
        image (numpy array): The input image in BGR or grayscale format.

    Returns:
        list: A list of segmented character images.
    """
    # Convert to grayscale if not already
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image

    # Apply Gaussian Blur to remove noise
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)

    # Apply Thresholding for binary image
    _, thresh = cv2.threshold(blurred, 150, 255, cv2.THRESH_BINARY_INV)

    # Find contours for segmentation
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    segmented_chars = []

    # Loop through contours to extract characters
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)

        # Filter small noise based on width and height
        if w > 10 and h > 10:
            char = thresh[y:y+h, x:x+w]

            # Resize character for better OCR (optional)
            char_resized = cv2.resize(char, (28, 28))

            segmented_chars.append(char_resized)

            # Optional: Display segmented character
            plt.imshow(char_resized, cmap='gray')
            plt.title("Segmented Character")
            plt.show()

    return segmented_chars
```

```python

import os
import cv2
import pandas as pd
import numpy as np

def load_data_with_segmentation(folder, data_dir):
    """
    Loads data and performs segmentation and OCR.

    Args:
        folder (str): Folder name containing 'labels.csv' and 'images/'.
        data_dir (str): Root directory containing the dataset.

    Returns:
        tuple: Arrays of images, labels, and OCR results.
    """
    csv_path = os.path.join(data_dir, folder, "labels.csv")
    image_folder = os.path.join(data_dir, folder, "images")

    df = pd.read_csv(csv_path)

    images = []
    labels = []
    ocr_results = []

    for index, row in df.iterrows():
        image_filename = row["IMAGE"]
        image_path = os.path.join(image_folder, image_filename)

        if os.path.exists(image_path):
            image = cv2.imread(image_path)
            segmented_chars = segment_characters(image)  # Use segmentation function

            ocr_text = ""

            # Apply OCR on each segmented character
            for char_img in segmented_chars:
                text = pytesseract.image_to_string(char_img, config='--psm 10')
                ocr_text += text.strip()

            images.append(image)
            labels.append(row["MEDICINE_NAME"])
            ocr_results.append(ocr_text)

            print(f"OCR Result for {image_filename}: {ocr_text}")

        else:
            print(f"⚠️ Warning: Image {image_filename} not found in {image_folder}!")

    return np.array(images), np.array(labels), ocr_results
```
