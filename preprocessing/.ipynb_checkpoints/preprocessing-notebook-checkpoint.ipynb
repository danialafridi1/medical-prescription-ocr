```python
import os
import pandas as pd
import numpy as np
import cv2
import matplotlib.pyplot as plt
from tensorflow.keras.utils import to_categorical
```

```python
# Define dataset paths
data_dir = "../dataset/"  # Update as needed
folders = ["Training", "Testing", "Validation"]
image_size = (250, 250)
```

```python
def load_data(folder):
    
    csv_path = os.path.join(data_dir, folder, "labels.csv")
    image_folder = os.path.join(data_dir, folder, "images")  # Ensure the correct folder

    # Load CSV file
    df = pd.read_csv(csv_path)

    images = []
    labels = []

    for index, row in df.iterrows():
        image_filename = row["IMAGE"]  # Image name from CSV
        image_path = os.path.join(image_folder, image_filename)  # Full image path
        
        if os.path.exists(image_path):  # Ensure image exists
            image = cv2.imread(image_path)
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for matplotlib
            image = cv2.resize(image, (224, 224))  # Resize images if needed
            
            images.append(image)
            labels.append(row["MEDICINE_NAME"])  # Assign label correctly
        else:
            print(f"⚠️ Warning: Image {image_filename} not found in {image_folder}!")

    return np.array(images), np.array(labels)
```

```python
train_images, train_labels = load_data("Training")
test_images, test_labels = load_data("Testing")
val_images, val_labels = load_data("Validation")
```

```python
# Display first image
plt.imshow(train_images[50])
plt.title(f"Label: {train_labels[50]}")
plt.axis("off")
plt.show()

# Convert labels to categorical (if needed)
unique_labels = list(set(train_labels))
label_to_index = {label: i for i, label in enumerate(unique_labels)}
train_labels = to_categorical([label_to_index[label] for label in train_labels], num_classes=len(unique_labels))
test_labels = to_categorical([label_to_index[label] for label in test_labels], num_classes=len(unique_labels))
val_labels = to_categorical([label_to_index[label] for label in val_labels], num_classes=len(unique_labels))

print("Preprocessing Complete! Data Ready for Training.")
```
